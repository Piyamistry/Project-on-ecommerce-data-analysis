## INTRODUCTION

E-commerce data analysis involves the examination and interpretation of data generated within the context of electronic commerce transactions and activities. In the modern digital era, e-commerce platforms generate vast amounts of data, including customer interactions, purchase behavior, website traffic, and product details. Analyzing this data can provide valuable insights for businesses to enhance their operations, improve customer experience, and make informed strategic decisions.or staying competitive and meeting customer expectations.

## Key Components of E-commerce Data Analysis:

Marketing Effectiveness:
Analyzing marketing campaigns, including email marketing, social media promotions, and online advertising, helps businesses understand their impact on customer acquisition and retention. Attribution modeling can be used to determine the contribution of each marketing channel to conversions.

### 3.Sales Analysis:

Compare sales performance across different channels or platforms.

E-commerce data analysis plays a critical role in identifying fraudulent activities and enhancing security measures. Machine learning models can be employed to detect anomalous transactions and patterns indicative of fraudulent behavior.

```python
conn=databaseConnect()
sql="SELECT * FROM customer"
customerFrame=pd.read_sql(sql,conn)
customerFrame.dropna(inplace=True)
customerFrame.drop_duplicates(inplace=True)

sql="SELECT * FROM product"
productFrame=pd.read_sql(sql,conn)
productFrame.dropna(inplace=True)
productFrame.drop_duplicates(inplace=True)

sql="SELECT * FROM order_details"
orderFrame=pd.read_sql(sql,conn)
orderFrame.dropna(inplace=True)
orderFrame.drop_duplicates(inplace=True)

print("After fixing data....")
print(customerFrame)
print(productFrame)
print(orderFrame)
```

## ● Identify the total number of customers city wise.

```python
customer_count_by_city = customerFrame['city'].value_counts()
plt.figure(figsize=(10, 6))
customer_count_by_city.plot(kind='bar', color=['skyblue','orange','magenta','green'])
plt.title('Total Number of Customers by City')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## ● Identify the most frequent customers based##  on their order history.

```python
customer_count = customerFrame.groupby('city').size().reset_index(name='count')
customer_count = customer_count.sort_values(by='count', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(customer_count['city'], customer_count['count'], color=['skyblue','yellow','red','cyan'])
plt.title('Number of Customers City-wise')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

## ● Determine the total number of products available by category.

```python
products_count_by_category = productFrame.groupby('category').size().reset_index(name='total_products')
products_count_by_category = products_count_by_category.sort_values(by='total_products', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(products_count_by_category['category'], products_count_by_category['total_products'], color=['skyblue','green','magenta'])
plt.title('Total Number of Products by Category')
plt.xlabel('Category')
plt.ylabel('Total Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## ● Calculate the average, maximum, and minimum prices for products.

```python
average_selling_price = productFrame['selling_price'].mean()
max_selling_price = productFrame['selling_price'].max()
min_selling_price = productFrame['selling_price'].min()
prices = [average_selling_price, max_selling_price, min_selling_price]
labels = ['Average Selling Price', 'Maximum Selling Price', 'Minimum Selling Price']
plt.figure(figsize=(8, 8))
plt.pie(prices, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Selling Prices')
plt.axis('equal')
plt.show()
```

## ● Calculate the top 10 orders product wise.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
orders_product_wise = merged_data.groupby('product_name')['quantity'].sum().reset_index()
top_10_orders = orders_product_wise.nlargest(10, 'quantity')
plt.figure(figsize=(12, 8))
plt.barh(top_10_orders['product_name'], top_10_orders['quantity'], color='skyblue')
plt.title('Top 10 Orders Product-wise')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

## ● Analyze the order status distribution (e.g., pending, delivered).

```python
order_status_distribution = orderFrame.groupby('order_status').size().reset_index(name='count')
order_status_distribution = order_status_distribution.sort_values(by='count', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(order_status_distribution['order_status'], order_status_distribution['count'], color=['magenta','cyan','darkgreen'])
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
```

## ● Calculate total revenue generated from orders product wise.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
merged_data['revenue'] = merged_data['quantity'] * merged_data['selling_price']
product_revenue = merged_data.groupby('product_name')['revenue'].sum().reset_index()
product_revenue = product_revenue.sort_values(by='revenue', ascending=False)
plt.figure(figsize=(12, 8))
plt.bar(product_revenue['product_name'], product_revenue['revenue'], color=['skyblue','orange','lightgreen','red','yellow','magenta'])
plt.title('Total Revenue Generated from Orders Product-wise')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue Generated')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

## ● Calculate total revenue product category wise percentage.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
merged_data['revenue'] = merged_data['quantity'] * merged_data['selling_price']
category_revenue = merged_data.groupby('category')['revenue'].sum().reset_index()
total_revenue = category_revenue['revenue'].sum()
category_revenue['percentage'] = (category_revenue['revenue'] / total_revenue) * 100
plt.figure(figsize=(8, 8))
plt.pie(category_revenue['percentage'], labels=category_revenue['category'], autopct='%1.1f%%', startangle=140)
plt.title('Total Revenue Product Category-wise Percentage')
plt.axis('equal')  
plt.show()
```

## ● Identify the most profitable products based on the difference between 
original and selling prices.


```python
productFrame['profit'] = productFrame['selling_price'] - productFrame['original_price']
most_profitable_products = productFrame.sort_values(by='profit', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(most_profitable_products['product_name'], most_profitable_products['profit'], color='skyblue')
plt.title('Top 10 Most Profitable Products')
plt.xlabel('Product Name')
plt.ylabel('Profit (Selling Price - Original Price)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## ● Analyze the performance of different product categories in terms of sales.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
category_sales = merged_data.groupby('category')['total_price'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(category_sales['category'], category_sales['total_price'], color='cyan')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## ● Identify top 10 slow-moving products based on low sales.

```python
product_sales = orderFrame.groupby('product_id')['total_price'].sum().reset_index()
slow_moving_products = product_sales.sort_values(by='total_price').head(10)
slow_moving_products = pd.merge(slow_moving_products, productFrame, on='product_id')
plt.figure(figsize=(12, 6))
plt.bar(slow_moving_products['product_name'], slow_moving_products['total_price'], color=['skyblue','orange','green','yellow','red','cyan','purple'])
plt.title('Top 10 Slow-Moving Products Based on Low Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## ● Identify the top 10 best-selling products.

```python
productFrame['profit'] = productFrame['original_price'] - productFrame['selling_price']
most_profitable_products = productFrame.sort_values(by='profit', ascending=False).head(10)  
plt.figure(figsize=(12, 8))
plt.bar(most_profitable_products['product_name'], most_profitable_products['profit'], color='magenta')
plt.title('Top 10 Most Profitable Products')
plt.xlabel('Product Name')
plt.ylabel('Profit (Original Price - Selling Price)')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

## ● Identify product names with the highest and lowest order quantities.

```python
product_order_quantities = orderFrame.groupby('product_id')['quantity'].sum().reset_index()
product_order_quantities = pd.merge(product_order_quantities, productFrame, on='product_id')
product_order_quantities = product_order_quantities.sort_values(by='quantity')
plt.figure(figsize=(12, 6))
plt.plot(product_order_quantities['product_name'], product_order_quantities['quantity'], marker='o', linestyle='-')
plt.title('Order Quantities for Each Product')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## ● Identify customers with the highest and lowest order quantities by 
customer name.

```python
customer_order_quantities = orderFrame.groupby('customer_id')['quantity'].sum().reset_index()
highest_order_quantities = customer_order_quantities.sort_values(by='quantity', ascending=False).head(10)
lowest_order_quantities = customer_order_quantities.sort_values(by='quantity', ascending=True).head(10)
fig, axes = plt.subplots(2, 1, figsize=(12, 14))
axes[0].bar(highest_order_quantities['customer_id'], highest_order_quantities['quantity'], color='skyblue')
axes[0].set_title('Customers with Highest Order Quantities')
axes[0].set_xlabel('Customer ID')
axes[0].set_ylabel('Total Quantity Ordered')
axes[1].bar(lowest_order_quantities['customer_id'], lowest_order_quantities['quantity'], color='skyblue')
axes[1].set_title('Customers with Lowest Order Quantities')
axes[1].set_xlabel('Customer ID')
axes[1].set_ylabel('Total Quantity Ordered')

plt.tight_layout()
plt.show()
```

## ● Determine the most preferred payment modes.

```python
payment_mode_distribution = orderFrame.groupby('payment_mode').size().reset_index(name='count')
payment_mode_distribution = payment_mode_distribution.sort_values(by='count', ascending=False)
plt.figure(figsize=(8, 8))
plt.pie(payment_mode_distribution['count'], labels=payment_mode_distribution['payment_mode'], autopct='%1.1f%%', startangle=140)
plt.title('Preferred Payment Modes')
plt.axis('equal')
plt.show()
```
